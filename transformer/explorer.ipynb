```python
import pandas as pd
from pathlib import Path
```

```python
base_dir = Path('food_data')


food_categories_code = ['0200', # Spices and Herbs
                        '0900', # Fruits and Fruit Juices
                        '1100', # Vegetables and Vegetable Products
                        '1200', # Nut and Seed Products
                        '1600', # Legumes and Legume Products
                        '2000'  # Cereal Grains and Pasta
                        ]


food_categories_id = [2, # Spices and Herbs
                      9, # Fruits and Fruit Juices
                      11, # Vegetables and Vegetable Products
                      12, # Nut and Seed Products
                      16, # Legumes and Legume Products
                      20  # Cereal Grains and Pasta
                     ]
```

```python
food_df = pd.read_csv(base_dir / 'food.csv')
food_df.head()
```

```python
food_df = food_df[food_df['data_type'] == 'foundation_food']
food_df = food_df[food_df['food_category_id'].isin(food_categories_id)]
food_df.head()
```

```python
hummus_fdc_id = 321358

n_lookup_df = pd.read_csv(base_dir / 'food_nutrient.csv')
n_lookup_df.head()
```

```python
nutrient_hummus = n_lookup_df[n_lookup_df['fdc_id'] == hummus_fdc_id]
nutrient_hummus.head()
```

```python
nutrients = pd.read_csv(base_dir / 'nutrient.csv')
print(len(nutrients))
nutrients.head()
```

```python
nutrients_base = [
    1008, # Energy (Kcal)
    1003, # Protein
    1004, # Total lipid (fat)
    1005, # Carbohydrate, by difference
    1050, # Carbohydrate, by summation
    1051, # Water
    1079, # Fiber, total dietary
    1087, # Calcium
    1090, # Magnesium
    1092, # Potassium
    1093, # Sodium
    1114, # Vitamin D
    1162, # Vitamin C
    1166, # Riboflavin
]
```

```python
base_nutrients = nutrients[nutrients['id'].isin(nutrients_base)]
base_nutrients
```

```python
hummus = nutrient_hummus[nutrient_hummus['nutrient_id'].isin(nutrients_base)]
hummus = hummus.sort_values('nutrient_id')[['amount', 'nutrient_id']]
hummus
```

```python
#base_nutrients['amount'] = nutrient_hummus['amount'].astype(float)
base = base_nutrients.merge(hummus, left_on='id', right_on='nutrient_id')
base[['name', 'amount', 'unit_name']]
```
